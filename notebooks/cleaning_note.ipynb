```python
columns_index= df.iloc[:, 6:]   

grouped = data.groupby(['country_id', 'product','product_code', 'flow', 'unit'])[columns_index.columns].sum().reset_index()
grouped.head(30)
```

OW WE WILL ANSWER THE FIRST QUESTION WHICH IS TO COMPARE volume vs value of trade and find the unit price

for most status , this was not provided, so we will need a query to only extract products from countries where the product has export volume (1000mt) and value(1000US$). and we will compare countries with similar products to see what value they are getting for the same product

|------------|---------|------|-------|
| Albania  | engineered wood products | IMPORTS | 15888.03/140.81 |

```python
grouped_melted = grouped.melt(id_vars=['country_id','product','product_code','flow','unit']).rename(columns={"variable":"year"})
grouped_melted = grouped_melted[grouped_melted['value'] != 0]
grouped_melted.head(10)

#grouped[(grouped['country_id']=="Albania") & (grouped['product']=="case materials") & (grouped['flow']=="EXPORTS")].iloc#[1,4]/ grouped[(grouped['country_id']=="Albania") & (grouped['product']=="case materials") & (grouped['flow']=="EXPORTS")].iloc[0,4]
```

```python
records = []
for country in grouped_melted['country_id'].unique():
    for product in grouped_melted[grouped_melted['country_id']==country]['product'].unique():
        for year in grouped_melted[(grouped_melted['country_id']==country) & (grouped_melted['product']==product)]['year'].unique():
            for flow in ["IMPORTS","EXPORTS"]:
                subset = grouped_melted[(grouped_melted['country_id']==country) & (grouped_melted['product']==product) & (grouped_melted['flow']==flow) & (grouped_melted['year']==year)]
                if subset.shape[0] == 2:
                    ratio = subset.iloc[0,-1] / subset.iloc[1,-1]
                    records.append({"country_id": country, "product": product, "flow": flow, "year": year, "ratio":round(ratio,2)})

unit_price=pd.DataFrame(records)

        
```

```python
def drop_rows(df, column_name, allowed_values):
    """
    Drops rows from a DataFrame if the values in a specified column
    are not found in the given list of allowed values.

    Args:
        df (pd.DataFrame): The DataFrame to modify.
        column_name (str): The name of the column to check for values.
        allowed_values (list): A list of allowed values to retain in the DataFrame.

    Returns:
        pd.DataFrame: The modified DataFrame with rows dropped.
    """
    return df[df[column_name].isin(allowed_values)]
```
